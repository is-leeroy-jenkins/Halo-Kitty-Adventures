## ðŸ§© Formula Fundamentals
- Formulas start with = and follow PEMDAS; use parentheses for clarity.
- Relative references (A1) shift; absolute ($A$1) remain fixed; mixed ($A1/A$1) lock one axis.
- Favor Named Ranges and Tables for readability.
- Wrap potential errors in IFERROR or IFNA.
- Limit volatile functions such as OFFSET, INDIRECT, RAND, NOW in large models.

```python
=A1 + B1 * C1
=$A1 * B$2
=SUM(Revenue)
=IFERROR(A1/B1, "Guarded division")
```

## ðŸ§® Math & Trigonometry
- Use SUM, PRODUCT, and SUMPRODUCT for totals and weighted KPIs.
- Apply ROUND, INT, and TRUNC to control display versus calculation precision.
- MOD manages cycles; POWER, SQRT, LOG, and LN handle growth or scale.
- Use SIN, COS, TAN families for engineering; convert with DEGREES or RADIANS.
- RAND and RANDBETWEEN are volatile; snapshot when reproducibility is required.

```python
=SUM(A1:A10)
=PRODUCT(B2:B5)
=ROUND(123.456,2)
=MOD(17,5)
=POWER(3,4)
=SUMPRODUCT(A2:A10,B2:B10)
=PI()
=SIN(PI()/2)
=LOG(100,10)
=RAND()
```

## ðŸ§  Logical
- IF and IFS for thresholds and multi-condition branching.
- SWITCH replaces nested IF statements for categorical lookups.
- Combine conditions with AND, OR, NOT, XOR.
- IFERROR and IFNA maintain model stability.
- TRUE and FALSE provide explicit logical constants.

```python
=IF(A1>100,"Over Budget","OK")
=IFS(A1>90,"A",A1>80,"B",TRUE,"F")
=SWITCH(D2,"N","North","S","South","Unknown")
=AND(A1>0,B1<100)
=IFERROR(A1/B1,"Division Error")
```

## ðŸ—“ï¸ Date & Time
- Dates are serial numbers; times are day fractions.
- Construct using DATE or TIME; extract with YEAR, MONTH, DAY, HOUR, MINUTE, SECOND.
- Use EDATE and EOMONTH for month offsets, WORKDAY and NETWORKDAYS for business calendars.
- DATEDIF and YEARFRAC compute durations and fractional years.
- Apply TEXT only for final display formatting.

```python
=TODAY()
=NOW()
=DATE(2025,11,6)
=YEAR(A1)
=MONTH(A1)
=DAY(A1)
=NETWORKDAYS(A1,B1)
=WORKDAY(A1,5)
=EOMONTH(A1,0)
=YEARFRAC(A1,B1)
```

## ðŸ”¤ Text
- Clean with TRIM and CLEAN; normalize case with UPPER, LOWER, PROPER.
- Slice using LEFT, RIGHT, MID; find with SEARCH or FIND.
- Replace with REPLACE or SUBSTITUTE.
- Combine with CONCAT or TEXTJOIN.
- Use TEXT for formatting at presentation time.
- TEXTSPLIT, TEXTBEFORE, and TEXTAFTER simplify tokenization.

```python
=LEN(A1)
=LEFT(A1,5)
=RIGHT(A1,3)
=SEARCH("cat",A1)
=SUBSTITUTE(A1,"old","new",2)
=TEXTJOIN(", ",TRUE,A1:A5)
=UPPER(A1)
=TEXT(A1,"mm/dd/yyyy")
=TEXTSPLIT(A1,",")
=TEXTAFTER(A1,"-")
```

## ðŸ” Lookup & Reference
- Prefer XLOOKUP for bi-directional searches with optional defaults.
- XMATCH supports wildcard and approximate matching.
- Combine INDEX and MATCH for 2D lookups.
- CHOOSECOLS and CHOOSEROWS project subsets.
- OFFSET and INDIRECT are volatile; use only for dynamic referencing.
- GETPIVOTDATA extracts PivotTable values safely.

```python
=XLOOKUP(E2,A2:A100,C2:C100,"Not Found")
=XMATCH("North",A2:A10,0)
=INDEX(C2:C10,MATCH(E2,A2:A10,0))
=CHOOSECOLS(A1:D10,1,4)
=GETPIVOTDATA("Sales",$A$3,"Region","West")
```

## ðŸ“Š Statistical
- Use AVERAGE, MEDIAN, and MODE for central tendency.
- COUNT, COUNTIF, and COUNTIFS for conditional counting.
- STDEV and VAR for spread; SKEW and KURT for distribution shape.
- PERCENTILE, QUARTILE, and RANK functions create performance bands.
- Correlation and regression analysis with CORREL, LINEST, and FORECAST.LINEAR.
- T.TEST, Z.TEST, and CHISQ.TEST provide quick hypothesis checks.

```python
=AVERAGE(A1:A10)
=COUNTIF(A1:A10,">100")
=STDEV.S(A1:A10)
=VAR.P(A1:A10)
=RANK.EQ(A1,A1:A10)
=PERCENTILE.INC(A1:A10,0.9)
=CORREL(A1:A10,B1:B10)
=FORECAST.LINEAR(2025,B2:B10,A2:A10)
=T.TEST(A1:A20,B1:B20,2,2)
```

## ðŸ’° Financial
- PMT, FV, and PV compute loan payments and value of money.
- RATE and NPER solve for interest rate or term.
- NPV/XNPV and IRR/XIRR assess investment returns.
- SLN, DDB, and DB calculate depreciation.
- PRICE and YIELD evaluate bonds; DURATION and MDURATION measure sensitivity.
- EFFECT and NOMINAL reconcile compounding rates.

```python
=PMT(0.05/12,60,-10000)
=FV(0.07,10,-1000)
=NPV(0.1,B2:B6)
=IRR(B2:B6)
=PRICE("1/1/2025","1/1/2030",0.05,0.052,100,2,0)
=YIELD("1/1/2025","1/1/2030",0.05,95,100,2,0)
=DURATION("1/1/2025","1/1/2030",0.05,0.052,2,0)
=EFFECT(0.05,12)
```

## ðŸ§± Database & Information
- DSUM, DCOUNT, and DAVERAGE treat ranges as databases with criteria ranges.
- CELL, INFO, and TYPE return metadata about cells and environment.
- IS functions check data types or errors.
- GETPIVOTDATA retrieves PivotTable measures deterministically.

```python
=DSUM(A1:E100,"Sales",G1:H2)
=DCOUNT(A1:E100,"ID",G1:H2)
=CELL("address",A1)
=ISNUMBER(A1)
=GETPIVOTDATA("Units",$A$3,"Region","West")
```

## ðŸ§¬ Dynamic Arrays & LAMBDA
- Arrays spill automatically; no Ctrl+Shift+Enter required.
- FILTER, SORT, and UNIQUE manage table subsets.
- CHOOSECOLS, CHOOSEROWS, VSTACK, and HSTACK reshape arrays.
- LET defines local variables; LAMBDA creates reusable custom functions.
- MAP, BYROW, and REDUCE perform higher-order transformations.

```python
=SEQUENCE(5)
=FILTER(A2:C100,C2:C100="West")
=SORT(A2:B10,2,-1)
=UNIQUE(A2:A100)
=LET(x,A1:A10,SUM(x)/COUNT(x))
=LAMBDA(a,b,a+b)(2,3)
```
