## ðŸ§© Formula Fundamentals

Excel formulas always begin with an equals sign `=`.
They can include **constants**, **cell references**, **operators**, and **functions**.
Evaluation follows the **PEMDAS** order: Parentheses â†’ Exponents â†’ Multiplication / Division â†’ Addition / Subtraction.

### Example â€” Operator Precedence

```excel
=A1 + B1 * C1
# Multiplies B1*C1 first, then adds A1
```

### Cell Reference Types

| Type     | Example        | Behavior            |
| -------- | -------------- | ------------------- |
| Relative | `A1`           | Changes when copied |
| Absolute | `$A$1`         | Remains fixed       |
| Mixed    | `$A1` or `A$1` | Locks row or column |

```excel
=$A1 * B$2
# Locks column A and row 2 respectively
```

### Named Ranges

Define names for ranges via **Formulas â†’ Name Manager**.

```excel
=SUM(Revenue)
# Sums named range "Revenue"
```

## ðŸ§® Math & Trigonometry Functions

### **ABS**

**Syntax:** `=ABS(number)`

```excel
=ABS(-23)
# Returns 23 â€“ absolute value of â€“23
```

### **SUM**

**Syntax:** `=SUM(number1,[number2], â€¦)`

```excel
=SUM(A1:A10)
# Adds all numbers in A1:A10
```

### **PRODUCT**

**Syntax:** `=PRODUCT(number1,[number2], â€¦)`

```excel
=PRODUCT(B2:B5)
# Multiplies all numbers in range B2:B5
```

### **ROUND / ROUNDUP / ROUNDDOWN**

**Syntax:** `=ROUND(number, num_digits)`

```excel
=ROUND(123.456, 2)
# Returns 123.46
=ROUNDUP(5.01, 0)  â†’ 6
=ROUNDDOWN(5.99, 0)  â†’ 5
```

### **INT / TRUNC**

**Syntax:** `=INT(number)`  `=TRUNC(number,[num_digits])`

```excel
=INT(7.8)  â†’ 7
=TRUNC(7.876, 2)  â†’ 7.87
```

### **MOD**

**Syntax:** `=MOD(number, divisor)`

```excel
=MOD(17, 5)
# Returns 2 (remainder of 17Ã·5)
```

### **POWER / SQRT**

```excel
=POWER(3, 4)  â†’ 81
=SQRT(16)  â†’ 4
```

### **RAND / RANDBETWEEN**

```excel
=RAND()
# Random decimal 0 â‰¤ x < 1
=RANDBETWEEN(1, 100)
# Random integer 1â€“100
```

### **SUMPRODUCT**

**Syntax:** `=SUMPRODUCT(array1,[array2], â€¦)`

```excel
=SUMPRODUCT(A2:A10, B2:B10)
# Multiplies each pair then sums the products
```

### **QUOTIENT**

**Syntax:** `=QUOTIENT(numerator, denominator)`

```excel
=QUOTIENT(10, 3)
# Returns integer portion of division â†’ 3
```

### **PI / EXP / LN / LOG / LOG10**

```excel
=PI()  â†’ 3.14159
=EXP(1)  â†’ e â‰ˆ 2.718
=LN(10)  â†’ 2.3026
=LOG(100, 10)  â†’ 2
=LOG10(1000)  â†’ 3
```

### **SIN / COS / TAN / ASIN / ACOS / ATAN**

```excel
=SIN(PI()/2)  â†’ 1
=COS(0)  â†’ 1
=ATAN(1)  â†’ 0.785 (radians)
```

### **DEGREES / RADIANS**

```excel
=DEGREES(PI())  â†’ 180
=RADIANS(90)  â†’ 1.571
```

## ðŸ§  Logical Functions

### **IF**

**Syntax:** `=IF(logical_test, value_if_true, value_if_false)`

```excel
=IF(A1>100, "Over Budget", "OK")
# Conditional evaluation
```

### **IFS**

**Syntax:** `=IFS(logical_test1,value_if_true1,[â€¦])`

```excel
=IFS(A1>90,"A",A1>80,"B",A1>70,"C",TRUE,"F")
# Multi-condition grader
```

### **SWITCH**

```excel
=SWITCH(D2,"N","North","S","South","Unknown")
# Maps codes to values without nested IFs
```

### **AND / OR / NOT / XOR**

```excel
=AND(A1>0, B1<100)
# TRUE if both true
=OR(C1="Y", C1="Yes")
# TRUE if either true
=NOT(A1>10)
# Logical negation
=XOR(A1>0, B1>0)
# TRUE if only one is true
```

### **IFERROR / IFNA**

```excel
=IFERROR(A1/B1,"Error Div/0")
# Handles errors gracefully
=IFNA(VLOOKUP(E2,A2:B100,2,0),"Not Found")
```

### **TRUE / FALSE**

```excel
=TRUE() â†’ TRUE
=FALSE() â†’ FALSE
```

## ðŸ—“ï¸ Date & Time Functions

Dates are stored as serial numbers (starting Jan 1 1900 = 1).
Times are fractions of a day ( 0.5 = 12:00 PM ).

### **TODAY / NOW**

```excel
=TODAY()
# Current date
=NOW()
# Current date and time
```

### **DATE / DATEVALUE**

```excel
=DATE(2025, 11, 6)
# Returns serial date for Nov 6 2025
=DATEVALUE("11/6/2025")
# Converts text to date
```

### **YEAR / MONTH / DAY / WEEKDAY / WEEKNUM**

```excel
=YEAR(A1)  â†’ 2025
=MONTH(A1)  â†’ 11
=DAY(A1)  â†’ 6
=WEEKDAY(A1, 2)  â†’ 1 (Monday = 1)
=WEEKNUM(A1, 2)  â†’ 45 (ISO week)
```

### **TIME / HOUR / MINUTE / SECOND**

```excel
=TIME(14, 30, 0)  â†’ 2:30 PM
=HOUR(NOW())  â†’ current hour
```

### **EDATE / EOMONTH / WORKDAY / NETWORKDAYS**

```excel
=EDATE(A1, 3)
# Adds 3 months
=EOMONTH(A1, 0)
# Last day of month
=WORKDAY(A1, 5)
# Date 5 workdays after A1
=NETWORKDAYS(A1,B1)
# Count of business days
```

### **DATEDIF / YEARFRAC**

```excel
=DATEDIF(A1,B1,"y")
# Whole years between dates
=YEARFRAC(A1,B1)
# Fractional year between dates
```

### **ISOWEEKNUM / TODAY-Based Calculations**

```excel
=ISOWEEKNUM(A1)
# ISO week number
=A1-TODAY()
# Days since/till today
```

## ðŸ”¤ Text Functions

### **LEN**

**Syntax:** `=LEN(text)`

```excel
=LEN(A1)
# Number of characters in A1
```

### **LEFT / RIGHT / MID**

```excel
=LEFT(A1, 5)
# First 5 characters
=RIGHT(A1, 3)
# Last 3 characters
=MID(A1, 4, 2)
# 2 chars starting at position 4
```

### **SEARCH / FIND**

```excel
=SEARCH("cat", A1)
# Position of "cat" (case-insensitive)
=FIND("cat", A1)
# Case-sensitive
```

### **REPLACE / SUBSTITUTE**

```excel
=REPLACE(A1, 1, 5, "Hello")
# Replace first 5 chars with "Hello"
=SUBSTITUTE(A1, "old", "new", 2)
# Replaces second occurrence
```

### **CONCAT / TEXTJOIN**

```excel
=CONCAT(A1," ",B1)
# Combine cells
=TEXTJOIN(", ", TRUE, A1:A5)
# Joins range with delimiter, ignoring blanks
```

### **UPPER / LOWER / PROPER / TRIM / CLEAN**

```excel
=UPPER(A1) â†’ "HELLO"
=LOWER(A1) â†’ "hello"
=PROPER(A1) â†’ "Hello"
=TRIM(A1)
# Removes extra spaces
=CLEAN(A1)
# Removes non-printable characters
```

### **TEXT**

```excel
=TEXT(A1,"mm/dd/yyyy")
# Format date as text
=TEXT(A1,"$#,##0.00")
# Numeric format string
```

### **VALUE**

```excel
=VALUE("123")
# Converts text to number
```

### **TEXTSPLIT / TEXTBEFORE / TEXTAFTER** (Excel 365+)

```excel
=TEXTSPLIT(A1,",")
# Splits text into cells by comma
=TEXTBEFORE(A1,"-")
# Text to left of first hyphen
=TEXTAFTER(A1,"-")
# Text to right of first hyphen
```

## ðŸ” Lookup & Reference Functions

### **VLOOKUP**

**Syntax:** `=VLOOKUP(lookup_value, table_array, col_index_num, [range_lookup])`

```excel
=VLOOKUP(E2, A2:D100, 3, FALSE)
# Finds E2 in first column of A2:D100 and returns matching 3rd-column value
```

### **HLOOKUP**

**Syntax:** `=HLOOKUP(lookup_value, table_array, row_index_num, [range_lookup])`

```excel
=HLOOKUP("Q1", A1:Z4, 3, FALSE)
# Searches first row and returns 3rd-row value
```

### **XLOOKUP**

**Syntax:** `=XLOOKUP(lookup_value, lookup_array, return_array, [if_not_found], [match_mode], [search_mode])`

```excel
=XLOOKUP(E2, A2:A100, C2:C100, "Not Found")
# Modern replacement for VLOOKUP/HLOOKUP
```

### **LOOKUP**

```excel
=LOOKUP(E2, A2:A100, C2:C100)
# Legacy approximate lookup
```

### **INDEX**

**Syntax:** `=INDEX(array, row_num, [column_num])`

```excel
=INDEX(B2:B10, 5)
# Returns 5th item in B2:B10
```

### **MATCH**

**Syntax:** `=MATCH(lookup_value, lookup_array, [match_type])`

```excel
=MATCH("North", A2:A10, 0)
# Position of â€œNorthâ€ in A2:A10
```

### **INDEX + MATCH Combination**

```excel
=INDEX(C2:C10, MATCH(E2, A2:A10, 0))
# Flexible lookup pair without column-index limitations
```

### **OFFSET**

**Syntax:** `=OFFSET(reference, rows, cols, [height], [width])`

```excel
=OFFSET(A1, 2, 1)
# Cell two rows down, one column right from A1
```

### **INDIRECT**

```excel
=INDIRECT("Sheet2!B3")
# Creates reference from text
```

### **CHOOSE / CHOOSECOLS / CHOOSEROWS**

```excel
=CHOOSE(2, "Red","Blue","Green")
# Returns "Blue"
=CHOOSECOLS(A1:D10, 1, 4)
# Returns only columns 1 and 4 from range
=CHOOSEROWS(A1:D10, 1, 3, 5)
# Returns rows 1, 3, and 5
```

### **COLUMN / ROW / COLUMNS / ROWS**

```excel
=COLUMN(B5) â†’ 2
=ROWS(A1:A10) â†’ 10
```

### **ADDRESS / INDIRECT with ADDRESS**

```excel
=ADDRESS(5, 2)
# Returns "$B$5"
=INDIRECT(ADDRESS(5, 2))
# Evaluates cell B5
```

### **TRANSPOSE**

```excel
=TRANSPOSE(A1:B3)
# Converts rows to columns (spills in 365)
```

### **GETPIVOTDATA**

```excel
=GETPIVOTDATA("Sales",$A$3,"Region","West")
# Extracts PivotTable value by field criteria
```

## ðŸ’° Financial Functions

### **PMT**

**Syntax:** `=PMT(rate, nper, pv, [fv], [type])`

```excel
=PMT(0.05/12, 60, -10000)
# Payment for 5% loan over 60 months
```

```excel
=FV(0.07, 10, -1000)
# Future value of annual $1000 payments @7%
=PV(0.05/12, 60, -200)
# Present value of $200 monthly payment
```

### **SLN / DDB / DB**

```excel
=SLN(10000, 1000, 5)
# Straight-line depreciation
=DDB(10000, 1000, 5, 1)
# Double-declining balance year 1
```

### **CUMIPMT / CUMPRINC**

```excel
=CUMIPMT(0.06/12, 60, 10000, 1, 12, 0)
# Total interest in year 1
```

### **EFFECT / NOMINAL**

```excel
=EFFECT(0.05, 12)
# Effective annual rate for monthly compounding
```

### **YIELD / PRICE / DURATION** *(bond analysis)*

```excel
=YIELD("1/1/2025","1/1/2030", 0.05, 95, 100, 2)
```

## ðŸ§¬ Dynamic Array and Lambda Functions (Excel 365 +)

### **SEQUENCE**

```excel
=SEQUENCE(5)
# {1;2;3;4;5}
=SEQUENCE(3, 4, 10, 5)
# 3Ã—4 matrix starting 10 increment 5
```

### **UNIQUE**

```excel
=UNIQUE(A2:A100)
# Distinct values only
```

### **FILTER**

```excel
=FILTER(A2:C100, C2:C100="West")
# Rows where Region="West"
```

### **SORT / SORTBY**

```excel
=SORT(A2:B10, 2, -1)
# Sort by column 2 descending
=SORTBY(A2:B10, B2:B10, 1)
# Sort based on another range
```

### **CHOOSECOLS / CHOOSEROWS**

*(already shown in Lookup section)*

### **VSTACK / HSTACK**

```excel
=VSTACK(A1:A3, B1:B3)
# Vertical append
=HSTACK(A1:C1, D1:F1)
# Horizontal append
```

### **WRAPCOLS / WRAPROWS**

```excel
=WRAPCOLS(A1:A12, 3)
# Wrap values into 3 columns
```

### **TAKE / DROP / EXPAND**

```excel
=TAKE(A1:D10, 5)
# First 5 rows
=DROP(A1:D10, -2)
# Remove last 2 rows
=EXPAND(A1:B2, 4, 3, "")
# Extend array with blanks
```

### **MAP / BYROW / BYCOL / REDUCE / SCAN**

```excel
=BYROW(A1:C3, LAMBDA(r, SUM(r)))
# Row-wise sum
=MAP(A1:A3, LAMBDA(x, x^2))
# Elementwise square
=REDUCE(0, A1:A5, LAMBDA(a,b,a+b))
# Accumulate sum
```

### **LET**

```excel
=LET(x, A1:A10, sum, SUM(x), count, COUNT(x), sum/count)
# Assigns variables within formula
```

### **LAMBDA**

```excel
=LAMBDA(x, x^2)(5)
# Returns 25
=LAMBDA(a,b,a+b)(2,3)
# Returns 5
```

### **MAKEARRAY**

```excel
=MAKEARRAY(3, 3, LAMBDA(r,c, r*c))
# 3Ã—3 multiplication table
```

### **RANDARRAY**

```excel
=RANDARRAY(5, 2, 0, 1, TRUE)
# 5Ã—2 array random decimals
```

## âš¡ Optimization and Best Practices

1. **Minimize volatile functions:** `OFFSET`, `INDIRECT`, `RAND`, `NOW`.
2. **Prefer Tables and Structured References** for scalable formulas.
3. **Use Dynamic Arrays** instead of CSE array formulas.
4. **Cache intermediate results** with `LET`.
5. **Label inputs** via Named Ranges for clarity.
6. **Avoid deep nesting** â€” use `IFS` or `SWITCH`.
7. **Profile with F9 evaluation** to inspect sub-expressions.
8. **Document key formulas** in adjacent comments or metadata sheet.
