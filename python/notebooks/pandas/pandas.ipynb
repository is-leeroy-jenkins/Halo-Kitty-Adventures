## ğŸ§¹ 4. Cleaning & Preparing Data

### Key Points



- ğŸ”„ Missing values with `fillna()` or `dropna()`

## ğŸ“ˆ 10. Time Series Features

### Key Points


- ğŸ•’ Convert to datetime via `to_datetime()`

- ğŸ—“ï¸ Set datetime index

- â• Resample using `.resample()`

- ğŸ§® Rolling windows

- ğŸ“‰ Use `.shift()` for lagged features

```python
ts = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Value': range(10)
})

ts = ts.set_index('Date')
ts['Rolling_3'] = ts['Value'].rolling(3).mean()
ts
```
