### ğŸ“Š Introduction to Matplotlib

Matplotlib is the foundational data visualization library in Python. It provides a flexible framework for creating static, animated, and interactive plots in a variety of formats. Designed to mimic MATLAB's plotting style, it is widely used for both quick plotting and full-fledged publication-quality figures.

- ğŸ”¹ Core library for data visualization in Python
- ğŸ”¹ Integrates with NumPy, Pandas, and SciPy
- ğŸ”¹ Produces line plots, bar charts, histograms, scatter plots, etc.
- ğŸ”¹ Supports object-oriented and pyplot APIs
- ğŸ”¹ Export to PNG, PDF, SVG, EPS formats

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

### ğŸ“ Plotting with Pyplot API

The `pyplot` interface mimics MATLAB and is best for quick and simple plotting. Each function modifies the current figure or axes.

- ğŸ”¹ `plot()` draws line charts
- ğŸ”¹ `scatter()` for individual points
- ğŸ”¹ `bar()` and `barh()` for bar charts
- ğŸ”¹ `hist()` for histograms
- ğŸ”¹ `imshow()` for image data

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.figure()
plt.plot(x, y, marker='o', linestyle='--', color='r')
plt.title("Line Chart Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

### ğŸ§± Object-Oriented API

For more complex layouts, use the object-oriented approach, which gives more control over plot components.

- ğŸ”¹ Explicit control via `Figure` and `Axes` objects
- ğŸ”¹ Ideal for multi-plot layouts
- ğŸ”¹ Better for reusable, maintainable code
- ğŸ”¹ Works well with interactive backends

```python
fig, ax = plt.subplots()
x = np.linspace(0, 2 * np.pi, 100)
ax.plot(x, np.sin(x))
ax.set_title("Sine Function")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
plt.show()
```

### ğŸ¨ Customization

Matplotlib allows nearly every aspect of the plot to be customized.

- ğŸ”¹ Change colors, markers, linestyles
- ğŸ”¹ Annotate text or labels
- ğŸ”¹ Add grids, legends, and tick formatting
- ğŸ”¹ Modify subplot spacing with `tight_layout()`
- ğŸ”¹ Use `rcParams` for global styles

```python
plt.plot(x, y, color='purple', linewidth=2, linestyle='-.')
plt.title("Customized Plot")
plt.grid(True)
plt.legend(["sin(x)"])
plt.annotate("Peak", xy=(1.57, 1), xytext=(3, 1.2),
             arrowprops=dict(facecolor='black'))
plt.tight_layout()
plt.show()
```

### ğŸ§® Working with Subplots

Use `subplots()` to organize multiple plots in one figure.

- ğŸ”¹ Create grid layouts of axes
- ğŸ”¹ Control figure size
- ğŸ”¹ Use axes methods to plot independently
- ğŸ”¹ Use `sharex` and `sharey` for synchronization

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
x = np.linspace(0, 10, 100)

axs[0, 0].plot(x, np.sin(x))
axs[0, 1].plot(x, np.cos(x))
axs[1, 0].plot(x, np.tan(x))
axs[1, 1].plot(x, -np.sin(x))

plt.tight_layout()
plt.show()
```

### ğŸŒˆ Colormaps and Heatmaps

Matplotlib supports colormaps for scalar fields, heatmaps, and image-like data.

- ğŸ”¹ Use `imshow()` or `pcolormesh()` for heatmaps
- ğŸ”¹ Apply `cmap` argument for different color schemes
- ğŸ”¹ Add colorbars with `colorbar()`
- ğŸ”¹ Good for correlation matrices, grids, etc.

```python
data = np.random.rand(10, 10)
plt.imshow(data, cmap='viridis')
plt.colorbar(label='Value')
plt.title("Heatmap")
plt.show()
```

### ğŸ’¾ Saving Figures

You can save plots directly to file using `savefig()`.

- ğŸ”¹ Supports PNG, PDF, SVG, EPS, TIFF
- ğŸ”¹ Call before `show()`
- ğŸ”¹ Adjust resolution using `dpi`
- ğŸ”¹ Transparent backgrounds supported

```python
plt.plot(x, y)
plt.title("Exported Plot")
plt.savefig("sine_plot.png", dpi=300, transparent=True)
plt.close()
```
